QUE SIGNIFICA CADA UNO
POP: Población total. Esta categoría probablemente representa el número total de personas en el conjunto de datos, independientemente de su estado laboral.

INAC: Inactivos. Esta categoría podría representar a personas que no están empleadas ni buscando empleo activamente en el momento de la recopilación de datos. Esto puede incluir a personas jubiladas, estudiantes a tiempo completo, personas que cuidan del hogar, etc.

ACT: Activos. Esta categoría probablemente representa a las personas que están empleadas o que están buscando activamente empleo en el momento de la recopilación de datos. Es decir, personas en la fuerza laboral activa.

EMP: Empleados. Esta categoría puede referirse específicamente a las personas que están empleadas en el momento de la recopilación de datos.

UNE: Desempleados. Esta categoría probablemente representa a las personas que están desempleadas y que están buscando activamente empleo en el momento de la recopilación de datos.

UNK: Desconocido. Esta categoría puede ser utilizada para personas cuyo estado laboral no pudo ser determinado o no está disponible en el conjunto de datos.
