```python
# NUEVA PRUEBA CON PYTRENDS, LA PRIMERA VEZ ME FUNCIONO!!!! al darle de segunda no 
from pytrends.request import TrendReq

# Inicializar objeto pytrends
pytrends = TrendReq(hl='en-US', tz=360)

# Buscar topics relacionados con una palabra clave
topics = pytrends.suggestions('Jobs')

# Imprimir los topics encontrados
print("Topics relacionados con 'jobs':")
for topic in topics:
    print(topic)

# Seleccionar el código del topic de interés
topic_code = topics[0]['mid']  # Seleccionamos el primer topic como ejemplo

# Realizar la consulta a Google Trends utilizando el código del topic
pytrends.build_payload(kw_list=[topic_code], timeframe='today 12-m', geo='US')
data = pytrends.interest_over_time()

# Imprimir los datos obtenidos
print("\nDatos de Google Trends para el topic seleccionado:")
print(data)
# esto es solo una prueba, pero no hemos puesto los rangos adecuados, ni de tiempo ni de pais
```
